```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('amount_of_people.csv')
print(data)
```

```python
cleanData = data[['Year', 'Amount of people']]
print(cleanData)
x = data[['Year']]
y = data[['Amount of people']]
```

```python
plt.title('Year vs amount of people', fontsize=18)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Amount of people', fontsize=10)
plt.xlim(1950, 2022)
plt.ylim(2500000000, 8000000000)
plt.scatter(x, y, alpha=0.3, color='green')
```

```python
regr = LinearRegression()
regr.fit(x, y)
print(f'Theta1: {regr.coef_[0][0]}')
print(f'Intercept: {regr.intercept_[0]}')
print(f'R: {regr.score(x, y)}')
predict = regr.predict(x)
```

```python
plt.title('Year vs amount of people', fontsize=18)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Amount of people', fontsize=10)
plt.xlim(1945, 2022)
plt.ylim(2400000000, 9000000000)
plt.scatter(x, y, alpha=0.3)
plt.plot(x, predict, color='red', linewidth=1)
plt.style.use('fivethirtyeight')
plt.show()
```
