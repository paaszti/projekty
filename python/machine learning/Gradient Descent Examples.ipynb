```python
import matplotlib.pyplot as plt
import numpy as np
```

$$f(x) = 2x^3 -7x^2 - 2x +7$$ 


```python
def f(x):
    return 2*x**3 - 7*x**2 - 2*x + 7
```

```python
x_1 = np.linspace(-5, 5, 1000)
plt.plot(x_1, f(x_1))
plt.xlim(-5, 5)
plt.ylim(-400, 50)
plt.show()
```

```python
def df(x):
    return 6*x**2-14*x-2
```

```python
def gradient_descent(der_func, ini_guess, multiplier=0.02, precission=0.0001):
    new_x = ini_guess
    x_list = [new_x]
    slope_list = [der_func(new_x)]
    
    for i in range(1000):
        prev_x = new_x
        gredient = der_func(prev_x)
        new_x = prev_x - multiplier * gredient
        
        step_size = abs(new_x - prev_x)
        x_list.append(new_x)
        slope_list.append(der_func(new_x))
        if step_size < precission:
            break
        return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(der_func=df, ini_guess=-10)
print(f'Local min occurs at: {local_min}')
print(f'Number of steps: {len(list_x)}')
```

```python
#Calling gradient descent function
local_min, list_x, deriv_list = gradient_descent(der_func=df, ini_guess=5)

#Plot function and derivative and scatter side by side
plt.figure(figsize=[15, 5])

#1 Chart - cost function
plt.subplot(1, 2, 1)#wiersze, kolumny, index wykresu

plt.xlim(-5, 5)
plt.ylim(-400, 50)

plt.plot(x_1, f(x_1), alpha=0.8)
plt.scatter(list_x, f(np.array(list_x)), color='red', s=100, alpha=0.5)

plt.title('Cost function', fontsize=20)
plt.xlabel('X', fontsize=12)
plt.ylabel('f(x)', fontsize=12)


#2 Chart - derivative(pochodna)
plt.subplot(1, 2, 2)

plt.ylim(-6, 8)
plt.xlim(-5, 5)

plt.plot(x_1, df(x_1), color='skyblue')
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)

plt.title('Slop of the cost function', fontsize=20)
plt.xlabel('X', fontsize=12)
plt.ylabel('df(x)', fontsize=12)

plt.grid()#kratka

plt.show()
```
